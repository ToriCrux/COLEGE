3- Crie uma função que faça a leitura de um arquivo .csv utilizando a biblioteca Pandas com os seguintes critérios:

*   Receber como o nome do arquivo;
*   Retornar um dataframe Pandas, contendo o conteúdo do arquivo passado.

```python
import pandas as pd

def ler_csv(nome_arquivo):
    try:
        df = pd.read_csv(nome_arquivo, dtype={'num_ra': str, 'id': str})
        return df
    except Exception as e:
        print(f"Ocorreu um erro ao ler o arquivo {nome_arquivo}: {e}")
        return None


if __name__ == "__main__":
    nome_arquivo_alunos = 'base_alunos.csv'
    nome_arquivo_livros = 'base_livros.csv'

    df_alunos = ler_csv(nome_arquivo_alunos)
    df_livros = ler_csv(nome_arquivo_livros)

    print("DataFrame de Alunos:")
    print(df_alunos)
    print("\nDataFrame de Livros:")
    print(df_livros)
```

4- Crie uma função que faça a busca do aluno:
* O parâmetro de entrada deverá ser o nome do aluno ou o seu ra.
* Deverá ter um parâmetro booleano chamado is_ra, com valor inicial igual a None, para indicar se é um nome ou um ra.
* Para busca por ra, o retorno deverá ser um arquivo (objeto) do tipo json
* Para busca por nome do aluno o retorno deverá ser um arquivo json tbm
            

```python
import pandas as pd
import json


def buscar_aluno(df, criterio, is_ra=None):
    if is_ra is None:
        is_ra = criterio.isdigit()

    if is_ra:
        criterio = str(criterio)
        resultado = df[df['num_ra'] == criterio]
        if resultado.empty:
            return json.dumps({"erro": "O RA de aluno pesquisado não se encontra na base."}, indent=4)
        aluno = resultado.iloc[0]

        aluno_dict = {
            "registro": int(aluno['num_ra']),
            "nome": aluno['nome'],
            "curso": aluno['curso'],
            "semestre": int(aluno['semestre']),
            "livros": {
                "livro_1": aluno['livro_1'],
                "livro_2": aluno['livro_2'],
                "livro_3": aluno['livro_3']
            }
        }
        return json.dumps(aluno_dict, indent=4)


    else:
        resultados = df[df['nome'].str.contains(criterio, case=False, na=False)]

        if resultados.empty:
            return json.dumps({"erro": "Aluno não encontrado"}, indent=4)
        alunos_dict = {
            "registros_encontrados": len(resultados),
            "alunos": {}
        }

        for index, aluno in resultados.iterrows():
            alunos_dict["alunos"][f"aluno_{index + 1}"] = {
                "registro": aluno['num_ra'],
                "nome": aluno['nome'],
                "curso": aluno['curso'],
                "semestre": aluno['semestre'],
                "livros": {
                    "livro_1": aluno['livro_1'],
                    "livro_2": aluno['livro_2'],
                    "livro_3": aluno['livro_3']
                }
            }

        return json.dumps(alunos_dict, indent=4)


if __name__ == "__main__":
    df_alunos = ler_csv('base_alunos.csv')

    if df_alunos is not None:
        while True:
            entrada = input("Digite o RA ou o nome do aluno (ou 'sair' para encerrar): ")
            if entrada.lower() == 'sair':
                break

            is_ra = None
            if entrada.isdigit():
                is_ra = True
            else:
                is_ra = False

            resultado = buscar_aluno(df_alunos, entrada, is_ra)
            print(resultado)

    else:
        print("Não foi possível carregar a base de alunos.")
```

5 - Crie uma função que realize o salvamento de um objeto json em um arquivo local obedecendo os seguintes critérios:
* Receber como parâmetros de entrada um objeto .json, o caminho onde será salvo o arquivo, chamado nome_arquivo, com valor inicial igual a None.
  * Caso o parâmetro nome_arquivo receba um valor válido, a função deve salvar o arquivo usando esse nome.
  * Caso o parâmetro nome_arquivo não seja passado (None), a função deverá gerar um nome de arquivo com base na data e hora do sistema, para ser salvo.

```python
import json
import os
from datetime import datetime

def salvar_json(objeto_json, nome_arquivo=None):
    if nome_arquivo is None:
        nome_arquivo = datetime.now().strftime("%Y-%m-%d_%H-%M-%S") + ".json"

    if not nome_arquivo.endswith('.json'):
        nome_arquivo += '.json'

    try:
        with open(nome_arquivo, 'w') as f:
            json.dump(objeto_json, f, indent=4)
        return os.path.abspath(nome_arquivo)
    except Exception as e:
        print(f"Erro ao salvar o arquivo JSON: {e}")
        return None

if __name__ == "__main__":
    objeto_json = {
        "nome": "João",
        "idade": 30,
        "cidade": "São Paulo"
    }

    caminho_arquivo = salvar_json(objeto_json, "meu_arquivo.json")
    if caminho_arquivo:
        print(f"Arquivo salvo com sucesso em: {caminho_arquivo}")
    else:
        print("Falha ao salvar o arquivo.")


    caminho_arquivo_auto = salvar_json(objeto_json)
    if caminho_arquivo_auto:
        print(f"Arquivo salvo com sucesso em: {caminho_arquivo_auto}")
    else:
        print("Falha ao salvar o arquivo.")
```

6 - Crie uma função que faça a busca do livro.
* O parâmetro de entrada deverá ser o nome do livro.
* Deverá ter um parâmetro booleano chamado is_id, com valor inicial igual a False, para indicar se é um nome ou um id.
* Para busca por id, o retorno deverá ser um arquivo (objeto) do tipo json
* Para a busca por nome do livro, o retorno deverá ser um objeto json


```python
import pandas as pd
import json

def buscar_livro(df, criterio, is_id=False):
    if is_id:
        criterio = str(criterio)
        resultado = df[df['id'] == criterio]

        if resultado.empty:
            return json.dumps({"erro": "O ID do livro pesquisado não se encontra na base."}, indent=4)

        livro = resultado.iloc[0]
        livro_dict = {
            "registro": str(livro['id']),
            "codigo": str(livro['codigo']),
            "titulo": livro['titulo'],
            "autores": [autor.strip() for autor in livro['autores'].split(',')],
            "volume": int(livro['volume']),
            "aluno": str(livro['aluno'])
        }

        return json.dumps(livro_dict, indent=4)


    else:
        resultados = df[df['titulo'].str.contains(criterio, case=False, na=False)]

        if resultados.empty:
            return json.dumps({"erro": "Livro não encontrado"}, indent=4)

        livros_dict = {
            "registros_encontrados": len(resultados),
            "livros": {}
        }

        for index, livro in resultados.iterrows():
            livros_dict["livros"][f"livro_{index + 1}"] = {
                "registro": livro['id'],
                "codigo": livro['codigo'],
                "titulo": livro['titulo'],
                "autores": [autor.strip() for autor in livro['autores'].split(',')],
                "volume": livro['volume'],
                "aluno": livro['aluno']
            }

        return json.dumps(livros_dict, indent=4)


if __name__ == "__main__":
    df_livros = ler_csv('base_livros.csv')

    if df_livros is not None:
        while True:
            entrada = input("Digite o nome ou id do livro (ou 'sair' para encerrar): ")
            if entrada.lower() == 'sair':
                break

            is_id = entrada.isdigit()

            resultado = buscar_livro(df_livros, entrada, is_id=is_id)
            print(resultado)
    else:
        print("Não foi possível carregar a base de livros.")
```
