```python
import sqlite3
from datetime import datetime, timedelta

# Conexão com o banco de dados SQLite
conn = sqlite3.connect('dados_empresa.db')
cursor = conn.cursor()

# Calcula a data de 3 meses atrás
data_3_meses_atras = datetime.now() - timedelta(days=90)
data_3_meses_atras_inicio = data_3_meses_atras.replace(day=1)  # Primeiro dia do mês há três meses atrás

# Consulta SQL para listar os funcionários que tiveram aumento salarial nos últimos 3 meses
query = '''
    SELECT DISTINCT h1.ID_FUNCIONARIO, f.NM_FUNCIONARIO
    FROM hist_salarios AS h1
    JOIN (
        SELECT ID_FUNCIONARIO, MIN(DT_RECEBIMENTO) AS DATA_MENOR
        FROM hist_salarios
        GROUP BY ID_FUNCIONARIO
    ) AS t ON h1.ID_FUNCIONARIO = t.ID_FUNCIONARIO
    JOIN hist_salarios AS h2 ON h1.ID_FUNCIONARIO = h2.ID_FUNCIONARIO
    JOIN funcionarios AS f ON h1.ID_FUNCIONARIO = f.ID_FUNCIONARIO
    WHERE h1.DT_RECEBIMENTO >= ?
    AND h1.VLR_SALARIO > h2.VLR_SALARIO
    AND h1.ID_HISTORICO != h2.ID_HISTORICO
'''

# Executa a consulta SQL
cursor.execute(query, (data_3_meses_atras_inicio.strftime('%d-%m-%Y'),))

# Obtém os resultados da consulta
results = cursor.fetchall()

# Imprimir os resultados
for row in results:
    print("\nID_FUNCIONARIO:", row[0])
    print("NM_FUNCIONARIO:", row[1])
    print("--------------")

# Fechar a conexão com o banco de dados
conn.close()
```

9- Listar qual departamento possui o maior número de dependentes

```python
conn = sqlite3.connect('dados_empresa.db')
cursor = conn.cursor()

query = '''
    SELECT
        D.NM_DEPARTAMENTO,
        COUNT(*) AS TOTAL_DEPENDENTES
    FROM
        DEPENDENTES DEP
    JOIN
        FUNCIONARIOS FN ON DEP.ID_FUNCIONARIO = FN.ID_FUNCIONARIO
    JOIN
        DEPARTAMENTOS D ON FN.ID_DEPARTAMENTO = D.ID_DEPARTAMENTO
    GROUP BY
        D.NM_DEPARTAMENTO
    ORDER BY
        TOTAL_DEPENDENTES DESC
    LIMIT 1
'''

cursor.execute(query)

result = cursor.fetchone()

if result:
    print("\nNM_DEPARTAMENTO:", result[0])
    print("TOTAL_DEPENDENTES:", result[1])
else:
    print("Não há resultados.")

conn.close()
```

10 - Listar a média de salário por departamento em ordem decrescente

```python
conn = sqlite3.connect('dados_empresa.db')
cursor = conn.cursor()

query = '''
    SELECT
        D.NM_DEPARTAMENTO,
        ROUND(AVG(H.VLR_SALARIO), 2) AS MEDIA_SALARIAL
    FROM
        HIST_SALARIOS H
    JOIN
        FUNCIONARIOS F ON H.ID_FUNCIONARIO = F.ID_FUNCIONARIO
    JOIN
        DEPARTAMENTOS D ON F.ID_DEPARTAMENTO = D.ID_DEPARTAMENTO
    GROUP BY
        D.NM_DEPARTAMENTO
    ORDER BY
        MEDIA_SALARIAL DESC
'''

cursor.execute(query)

results = cursor.fetchall()

for row in results:
    print("\nNM_DEPARTAMENTO:", row[0])
    print("MEDIA_SALARIAL:", row[1])

conn.close()
```
